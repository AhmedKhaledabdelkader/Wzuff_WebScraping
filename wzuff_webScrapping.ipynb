### Import modules

```python
import requests
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
```

### link chrome driver to selenium

```python
s=Service(r"H:\chromedriver-win64\chromedriver.exe")

driver=webdriver.Chrome(service=s)
```

### pass wzuff url to selenium

```python
wzuff_url="https://wuzzuf.net/jobs/egypt" 

driver.get(wzuff_url)

time.sleep(3)
```

### find and enter my input

```python
search_input=input("please enter your job search")

search= driver.find_element(By.XPATH,"""/html/body/div[1]/div/div/main/div[2]/div[1]/form/div/input""")

search.send_keys(search_input)
search.send_keys(Keys.ENTER)
```

### current url of driver

```python
driver_url=driver.current_url
driver_url
```

### get the url by the beautiful soup

```python
response=requests.get(driver_url)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
soup
```

### get all jobs

```python
all_jobs=soup.find_all("div",class_="css-1gatmva e1v1l3u10")
all_jobs
```

### get the first job

```python
my_job=soup.find("div",class_="css-1gatmva e1v1l3u10")
my_job
```

### job title of the first job

```python
job_title=my_job.find("h2",class_="css-m604qf")
print(job_title.text)
```

### job link of the first job

```python
job_link=job_title.a.get("href")
print(job_link)
```

### job company of the first job

```python
job_company=my_job.find("div",class_="css-d7j1kk").a.text.strip('-')
print(job_company)
```

### job location of the first job

```python
job_location=my_job.find("div",class_="css-d7j1kk").find("span",class_="css-5wys0k").text
print(job_location)
```

### skills of the first job

```python
s2=Service(r"H:\chromedriver-win64\chromedriver.exe")
driver2=webdriver.Chrome(service=s2)
driver2.get(job_link)
time.sleep(2)
my_html=driver2.page_source
soup2=BeautifulSoup(my_html,"html.parser")
skills_and_tools=soup2.find("div",class_="css-s2o0yh")
skillsList=skills_and_tools.find_all("a",class_="css-g65o95")
skills=[skill.text for skill in skillsList]
print(skills)
driver2.quit()
```

### get jobs in all pages

```python
data=[]
check=False
pagination_line=soup.find("li",class_="css-8neukt")
if pagination_line==None:
    jobs=soup.find_all("div",class_="css-1gatmva e1v1l3u10")
    no_of_jobs=len(jobs)
    pagination_line=f"0 of {no_of_jobs} "
    check=True

if not check:
    pagination_line=pagination_line.text



    
total_no_of_jobs=pagination_line[pagination_line.find("of ")+3:]
print(total_no_of_jobs)
page=0
while True:
    my_response=requests.get(driver_url+f"&start={page}")
    time.sleep(2)
    my_soup=BeautifulSoup(my_response.text,"html.parser")
    jobs=my_soup.find_all("div",class_="css-1gatmva e1v1l3u10")
    for job in jobs:
        job_title=job.find("h2",class_="css-m604qf")
        job_title_text=job_title.text
        job_link=job_title.a.get("href")
        job_company=job.find("div",class_="css-d7j1kk").a.text.strip('-')
        job_location=job.find("div",class_="css-d7j1kk").find("span",class_="css-5wys0k").text
        #............skills............#
        chrome_driver_path = r"H:\chromedriver-win64\chromedriver.exe"  
        chrome_options = Options()
        chrome_options.add_argument("--headless")  
        chrome_options.add_argument("--disable-gpu")
        chrome_service = Service(executable_path=chrome_driver_path)
        driver2 = webdriver.Chrome(service=chrome_service, options=chrome_options)
        driver2.get(job_link)
        time.sleep(4)
        my_html=driver2.page_source
        soup2=BeautifulSoup(my_html,"html.parser")
        skills_and_tools=soup2.find("div",class_="css-s2o0yh")
        skillsList=skills_and_tools.find_all("a",class_="css-g65o95")
        skills=[skill.text for skill in skillsList]
        driver2.quit()
        skills_and_tools=",".join(skills)
        
        job_data={"Job_Name":job_title_text,"Job_Link":job_link,"Job_Company":job_company,"Job_Location":job_location,"Job_Skills":skills_and_tools}
        
        data.append(job_data)
        
   
    
    
    if int(total_no_of_jobs)//15==page: # stopping critria for multipages
        break
    page+=1
print(page)   
```

### make data in csv file

```python
with open("wzuffScrapping.csv","w",newline="") as file:

    writer=csv.DictWriter(file,fieldnames=["Job_Name","Job_Link","Job_Company","Job_Location","Job_Skills"])
    
    writer.writeheader()

    writer.writerows(data)

    print("csv file is maken")
```
